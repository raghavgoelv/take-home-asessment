# Importing Libraries

```python
#importing required libraries

import pandas as pd
import json
import io
from pandas import json_normalize
from sqlalchemy import create_engine


pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

# Load Data & Converting JSON to Dataframe

```python
#users json

user_data=[]
user_file = open("users.json",'r')
for line in user_file:
    json_user_data = json.loads(line)
    user_data.append(json_user_data)

user_df = pd.json_normalize(user_data)
user_df.rename(columns={'_id.$oid':'userId', "createdDate.$date":'createdDate',
                        "lastLogin.$date":"lastLogin"},inplace=True)

user_df = user_df[['userId'] + [col for col in user_df.columns if col != 'userId']]

#date formats
date_cols = ['createdDate', 'lastLogin']

for col in date_cols :
    user_df[col] = pd.to_datetime(user_df[col], unit='ms')



print(user_df.shape)
display(user_df.head(5))
```

```python
#brands json

brand_data=[]
brand_file = open("brands.json",'r')
for line in brand_file:
    json_brand_data = json.loads(line)
    brand_data.append(json_brand_data)

brand_df = pd.json_normalize(brand_data)
brand_df.rename(columns={'_id.$oid':'brandId',"cpg.$id.$oid":"cpgId","cpg.$ref":"cpg_ref"},inplace=True)

brand_df = brand_df[['brandId'] + [col for col in brand_df.columns if col != 'brandId']]



print(brand_df.shape)
display(brand_df.head(5))
```

```python
 #receipts json

receipt_data=[]
receipt_file = open("receipts.json",'r')
for line in receipt_file:
    json_receipt_data = json.loads(line)
    receipt_data.append(json_receipt_data)

receipt_df = pd.json_normalize(receipt_data)
receipt_df.rename(columns={'_id.$oid':'receiptId',"createDate.$date":'createDate',
                          "dateScanned.$date":'dateScanned', 'finishedDate.$date':'finishedDate',
                          'modifyDate.$date':'modifyDate','pointsAwardedDate.$date':'pointsAwardedDate',
                          'purchaseDate.$date':'purchaseDate'},inplace=True)

receipt_df = receipt_df[['receiptId'] + [col for col in receipt_df.columns if col != 'receiptId']]

#date formats
date_cols = ['createDate', 'dateScanned', 'finishedDate', 'modifyDate', 'pointsAwardedDate', 'purchaseDate']

for col in date_cols :
    receipt_df[col] = pd.to_datetime(receipt_df[col], unit='ms')

print(receipt_df.shape)
display(receipt_df.head(5))
```

```python
rewards_list_df = receipt_df[['receiptId','rewardsReceiptItemList']]
rewards_list_df['rewardsReceiptItemList'].fillna(value=pd.Series([[]]*len(rewards_list_df)), inplace=True)

normalized_data = []
for idx, row in rewards_list_df.iterrows():
    id_value = row['receiptId']
    rewards_list = row['rewardsReceiptItemList']
    if rewards_list:  # Check if rewards_list is not empty
        normalized = json_normalize(rewards_list)
        normalized['receiptId'] = id_value
        normalized_data.append(normalized)

rewards_item_list_df = pd.concat(normalized_data, ignore_index=True)
rewards_item_list_df = rewards_item_list_df[['receiptId'] + 
                                            [col for col in rewards_item_list_df.columns if col != 'receiptId']]

receipt_df.drop(columns = ['rewardsReceiptItemList'], inplace = True)
print(rewards_item_list_df.shape)
display(rewards_item_list_df.head(5))
```
